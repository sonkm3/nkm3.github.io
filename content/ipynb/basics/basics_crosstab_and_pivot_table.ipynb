```python
import io
import pandas as pd

csv = '''date,area,count
2021-04-01, area1, 5
2021-04-01, area2, 3
2021-04-01, area3, 9
2021-04-02, area1, 4
2021-04-02, area2, 6
2021-04-02, area3, 2
2021-04-03, area1, 8
2021-04-03, area2, 4
2021-04-03, area3, 7
2021-04-04, area1, 1
2021-04-04, area2, 4
2021-04-04, area3, 8
2021-04-05, area1, 8
2021-04-05, area2, 1
2021-04-05, area3, 4
'''

data = pd.read_csv(io.StringIO(csv), parse_dates = ['date', ], low_memory=False)
data
```

```python
# ピボットテーブルをpivot_tableで作る
pivot_data = pd.pivot_table(data, index='date', columns='area', margins=True, aggfunc=sum)
pivot_data
```

```python
# crosstabでpivot_tableと同じようなピボットテーブルを作る
cross_data = pd.crosstab(data['date'], data['area'], values=data['count'], aggfunc=sum, margins=True)
cross_data
```

```python
# crosstabはnormalize引数を指定すると結果を行方向、列方向に正規化できる
cross_data = pd.crosstab(data['date'], data['area'], values=data['count'], aggfunc=sum, normalize='index')
cross_data
```
